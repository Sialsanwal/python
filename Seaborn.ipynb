# Correlation with Heatmap
- A correlation heatmap uses colored cells typically in a monochromatic scale, to show a 2D correlation matric(table) between two discreate dimesnions or events type . Its is very important in features selection

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

# JoinPlot
- A join plot allow to study the relationship between 2 numerical variable. The central chart dispalay their correlation . It is ususally a scatterplot ,  a hexbin plot , a 2D histrogram or a 2D density plot

```python
sns.jointplot(x='tip',y='total_bill',data=df)
```

```python
sns.jointplot(x='tip',y='total_bill',data=df, kind='reg')
```

```python
sns.pairplot(df)
```

```python
df['sex'].value_counts()
```

```python
sns.pairplot(df,hue='sex')
```

```python
df['smoker'].value_counts()
```

```python
sns.pairplot(df,hue='smoker')
```

```python
sns.displot(df['tip'])
```

# Box Plot
- A box and whisker plot(sometime called a box_ploy) is a graph that represent information from five-number summary

```python
sns.boxplot(x="smoker", y="total_bill", data=df)
```

```python
sns.boxplot(x="sex",y="total_bill", data=df)
```

```python
sns.boxplot(x="day",y="total_bill", data=df)
```

```python
sns.boxplot(x="day",y="total_bill", data=df, palette="rainbow")
```

```python
sns.boxplot(data=df)
```

```python
# Category my data based on some other categories
sns.boxplot(x="total_bill", y="day", hue="smoker", data=df)
```

# Violin Plot
- Violon plot helps us to see both the distribution of data in terms of kernal density estimation and box plot

```python
sns.violinplot("total_bill","day", data=df, palette="rainbow")
```
