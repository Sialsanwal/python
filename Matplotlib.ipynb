# MatplotLib 
- Matplotlib is a plotting library for the Python Programming language and its numerical mathematics extension NumPy. It provide an Object-Orinted API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, WxPython , QT or GTK+
- Genrally easy to get started for simple plots
- Support for custom labels and texts
- Great control of every elememnt in the figure 
- High-quality output in many format
- very customizable in general

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
```

```python
## ploting using matplotlib
## plt scatter
plt.scatter(x,y, c="r")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("Graph in 2-D")
plt.savefig("test.png")
```

```python
y=x*x
```

```python
#plt plot
plt.plot(x,y,'g--')
```

```python
#plt plot
plt.plot(x,y,'g-*')
```

```python
#plt plot
plt.plot(x,y,'g-+')
```

```python
#plt plot
plt.plot(x,y,'r',linestyle='dashed',linewidth=3, markersize=12)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("2D- Graph")
```

```python
# creating subplot
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'bo')
plt.subplot(2,2,3)
plt.plot(x,y,'y*--')
plt.subplot(2,2,4)
plt.plot(x,y,'go')
```

```python
x=np.arange(1,11)
y=3*x +5
plt.title("Matplotlib Demo")
plt.xlabel("X-axis label")
plt.ylabel("Y-axis label")
plt.plot(x,y, 'bo')
plt.show()
```
